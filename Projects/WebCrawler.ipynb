```python
import requests
from bs4 import BeautifulSoup
```

```python
class Spider:
    """The main class of the webcrawler."""
    
    def __init__(self):
        
        string = """
           _____       _     _           
          / ____|     (_)   | |          
         | (___  _ __  _  __| | ___ _ __ 
          \___ \| '_ \| |/ _` |/ _ \ '__|
          ____) | |_) | | (_| |  __/ |   
         |_____/| .__/|_|\__,_|\___|_|   
                | |                      
                |_|                      
        """                 
        
        print(string)

    web_directories = []

    def crawl(self, url):
        Spider.web_directories = self.crawling(url)
        active_dirs = Spider.web_directories

        for i in range(10):
            """Iteration loop."""
            
            new_dirs = []
            for j, web_directory in enumerate(active_dirs):
                """Crawling each link in the list of directories."""

                crawled_links = self.crawling(url + web_directory)
                Spider.web_directories.extend(crawled_links)
                list_length = len(Spider.web_directories)

                if not crawled_links:
                    pass
                else: 
                    new_dirs.extend(crawled_links)

                b = "[i={} | dirs: {} | {}/{}]".format(i + 1, list_length, j, len(active_dirs))
                print("\r", b, end = " ")

            active_dirs = new_dirs

        return Spider.web_directories

    def crawling(self, url):
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
        links_found = []
        response = requests.get(url, headers = headers)
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Seaching for links from page source,
        for link in soup.find_all("a"):
            potential_link = link.get("href")

            try:
                if potential_link[0] == "/" and bool(potential_link not in Spider.web_directories) == True:
                    links_found.append(potential_link)
            except Exception as error:
                pass

        return links_found
```

```python
CRAWLER = Spider()
CRAWLER.crawl("https://80000hours.org/career-guide/introduction/")
```
